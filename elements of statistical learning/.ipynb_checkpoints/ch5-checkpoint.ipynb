```python
class Logistic:
    def __init__(self,betas=None,p=None,K=None,covs=None):
        self.betas =betas
        self.p =p
        self.K =K
        
        self.covs =covs
        
    
    def logistic(self,x,betas,k):
        deno =np.exp(np.dot(betas[:,k],x))
        nume1 =betas.T.dot(x[:,np.newaxis])
        nume2 =np.exp(nume1[:,0])
        nume =1+(nume2.sum())
        res =(deno/nume)
        
        #np.exp()可以向量化计算吗，我下面是把其当作矩阵化算不合适
        #res =((np.exp(np.dot(betas[:,k],x)))/(1+\
        #    np.sum(np.exp((betas.T.dot(x[:,np.newaxis]))[:,0])))
        #print('res ',res)
        
        #res 变成0 这里就不好算了，增加eps看数值稳定会怎样
        eps =0.0000000001
            
        return res+eps
        
    
   
        
      
    
    def fit(self,X,y,tol=0.0001):
        Xs =X.copy()
        ys =y.copy()
        [N,p] =Xs.shape
        Xs =np.concatenate([np.ones((N,1),dtype=float),Xs],axis=1)
        K =ys.shape[1]
        ys =np.argmax(ys,axis=1)
        
        
            
        
        betas_last =np.random.randn(p+1,K-1)
        betas_cur  =np.zeros((p+1,K-1),dtype=float)
        k =-1
        time =0
        while(1):
            
            if(np.linalg.norm(betas_last-betas_cur)<tol):
                break
            betas_last =betas_cur.copy()
            k =(k+1)%(K-1)
            time =time+1
            #print('time ',time)
            yk =(ys==k)
            beta_last =np.random.randn(p+1)
            beta_cur  =betas_cur[:,k].copy()
            
            step =0
            
            while(1):
                if(np.linalg.norm(beta_last-beta_cur)<tol):
                    break
                
                step =step+1
                beta_last =beta_cur.copy()
                
                P=np.array([self.logistic(x,betas_cur,k) for x in Xs])

                W =np.diag(np.multiply(P,1.0-P))
                beta_cur =beta_last +np.linalg.inv(Xs.T.dot(W).dot(Xs)).dot(Xs.T).dot(yk[:,np.newaxis]-P[:,np.newaxis])[:,0]
                betas_cur[:,k]=beta_cur
        
        
        covs =np.zeros((K-1,p+1,p+1),dtype=float)
        for k in range(K-1):
            P =np.array([self.logistic(x,betas_cur,k) for x in Xs])
            W =np.diag(np.multiply(P,1.0-P))
            covs[k]=np.linalg.inv(Xs.T.dot(W).dot(Xs))
        
        
        self.covs =covs
        self.p =p
        self.K =K
        self.betas =betas_cur
        
            
        
        
        
        
    def predict(self,X,prob=0):
        Xs =X.copy()
        [N,p] =Xs.shape
        K =self.K
        Xs =np.concatenate([np.ones((N,1),dtype=float),Xs],axis=1)
        
        y_predict =np.zeros((N,K),dtype=float)
        
        for k in range(K-1):
            y_predict[:,k] =np.array([self.logistic(x,self.betas,k) for x in Xs],dtype=float)
        y_predict[:,K-1]=1.0- np.sum(y_predict[:,:K-1],axis=1)
        
        if(prob==1):
            return y_predict
        
        y_predict =np.argmax(y_predict,axis=1)
        return y_predict
    
    def residual_deviance(self,X,y):
        Xs =X.copy()
        ys =y.copy()
        [N,p]=Xs.shape
        #print('X shape ',Xs.shape)
        lgs =Logistic()
        lgs.fit(Xs,ys,tol=0.0001)
        y_predict =lgs.predict(Xs,prob=1)
        indexs =np.argmax(ys,axis=1)
        res =0.0
        for i in range(N):
            res =res + 2*np.log(y_predict[i,indexs[i]])
        return -1*res
        
        
    def AIC(self,X,y):
        Xs =X.copy()
        ys =y.copy()
        [N,p]=Xs.shape
        return self.residual_deviance(Xs,ys)+2*(p+1)
    
    def analysis_of_deviance(self,X,y,mask,num=0):
        Xs =X.copy()
        ys =y.copy()
        [N,p]=Xs.shape
        #mask =np.ones((p,),dtype=int)
        drop =False
        index =-1
        indexs =np.array(range(p),dtype=int)
        full_deviance =self.residual_deviance(Xs[:,indexs[mask!=0]],ys)
        nonzeros =np.count_nonzero(mask)
        if(nonzeros==num):
            return [mask,full_deviance]
        
        print('mask ',mask)
        print('full_deviance ',full_deviance)
        min_val =9999999999.0  
        #H0:theta=0,H1:theta!=0  score 最小的是最可能H0满足，是需要丢掉的特征
        #特征个数需要通过CV确定
        for i in range(p):
            if(mask[i]==0):
                continue
            mask[i]=0
            deviance =self.residual_deviance(Xs[:,indexs[mask!=0]],ys)
            if(min_val>deviance):
                index =i
                min_val =deviance    
            mask[i]=1
            
        
        
        mask[index]=0
        print('min_val ',min_val)
        print('mask ',mask)
        return self.analysis_of_deviance(Xs,ys,mask,num)
        
        
```

```python
class NatureSpline:
    
    def __init__(self,K=None):
        self.K=K
    
    def d(self,X,k,knots):
        Xs =X.copy()
        N=Xs.shape[0]
        res =np.zeros((N,))
        for i in range(N):
            x =Xs[i]
            a=np.power(x-knots[k],3)
            b=np.power(x-knots[-1],3)
            #print('a ',a,' b ',b)
            a=(int(a>=0))*(a)
            b=(int(b>=0))*(b)
            #print('a ',a,'b ',b)
            #res[i]=(1.0*a -b)/(1.0*knots[-1]-1.0*knots[k])
            res[i]=(a -b)/(knots[-1]-knots[k])
            #print('res[i] ',res[i])
        return res
    
    
    def basis(self,X,l):
        Xs =X.copy()
        if(l==1):
            return Xs.reshape(Xs.shape[0],1)
        N=Xs.shape[0]
        X_data =np.zeros((N,l),dtype=float)
        pers=np.array([0,25,50,75,100],dtype=float)
        knots =np.percentile(Xs,pers)
        #print('knots ',knots)
        X_data[:,0]=Xs
        dd =np.zeros((N,4),dtype=float)
        for i in range(4):
            dd[:,i] =self.d(Xs,i,knots)
        #print('dd ',dd)
        for i in range(4-1):
            X_data[:,i+1]=dd[:,i]-dd[:,3]
        #print('X_data ',X_data)
        return X_data
        
        
    def generate(self,X,lens,K=4):
        Xs =X.copy()
        [N,p]=Xs.shape
        indexs =np.array(range(p),dtype=int)
        mask =np.ones((p,),dtype=int)
        X_data =np.zeros((N,np.sum(lens)),dtype=int)
        start =0
        for i in range(p):
            print('start ',start)
            print('len ',lens[i])
            X_data[:,start:(start+lens[i])] =self.basis(Xs[:,i],lens[i])
            start =start+lens[i]
        return X_data
    
            
    
```

```python
def stepwise_select(lgs,X,y,mask,indexs,lens,num):
    Xs =X.copy()
    ys =y.copy()
    #print('num ',num)
    [N,p] =Xs.shape
    masks =np.ones((np.sum(lens),),dtype=int)
    for i in range(mask.shape[0]):
        if(mask[i]==0):
            masks[np.sum(lens[:i]):(np.sum(lens[:i])+lens[i])]=0
    
    aic =lgs.AIC(Xs[:,indexs[masks!=0]],ys)
    deviance =lgs.residual_deviance(Xs[:,indexs[masks!=0]],ys)
    #print('mask ',mask)
    #print('aic ',aic)
    #print('deviance ',lgs.residual_deviance(Xs[:,indexs[masks!=0]],ys))
    min_val =999999.0
    deviance2=0.0
    index =-1
    nonzeros =np.count_nonzero(mask)
    
    if(nonzeros<=num):
        return [mask,aic,deviance]
    
    for i in range(mask.shape[0]):
        if(mask[i]==0):
            continue
        else:
            mask[i]=0
            masks[np.sum(lens[:i]):(np.sum(lens[:i])+lens[i])]=0
            
            aic2 =lgs.AIC(Xs[:,indexs[masks!=0]],ys)
            #print('mask ',mask)
            #print('aic2 ',aic2)
            #print('deviance ',lgs.residual_deviance(Xs[:,indexs[masks!=0]],ys))
            if(aic2<min_val):
                min_val =aic2
                index =i
            masks[np.sum(lens[:i]):(np.sum(lens[:i])+lens[i])]=1
            mask[i]=1
    
    if(min_val>aic):
        return [mask,aic,deviance]
    else:
        mask[index]=0
        masks[np.sum(lens[:index]):(np.sum(lens[:index])+lens[index])]=0
        return stepwise_select(lgs,Xs,y,mask,indexs,lens,num)
            
    
```

```python
def get_vars(X,lens,beta,cov):
    Xs =X.copy()
    [N,p]=Xs.shape
    m =lens.shape[0]
    vals =np.zeros((N,m),dtype=float)
    variances =np.zeros((N,m),dtype=float)
    index =0
    for i in range(m):
        if(i==0):
            print('begin 0',' end ',lens[i]-1)
            X_choice =Xs[:,:lens[i]]
            beta_choice =beta[:lens[i]][:,np.newaxis]
            cov_choice =cov[:,:lens[i]]
            cov_choice =cov_choice[:lens[i],:]
        else:
            print('begin ',lens[i-1],' end ',lens[i]-1)
            X_choice =Xs[:,lens[i-1]:lens[i]]
            beta_choice =beta[lens[i-1]:lens[i]][:,np.newaxis]
            cov_choice =cov[:,lens[i-1]:lens[i]]
            cov_choice =cov_choice[lens[i-1]:lens[i],:]
        print('X_choice.shape ',X_choice.shape)
        print('beta_choice.shape ',beta_choice.shape)
        print('cov_choice.shape ',cov_choice.shape)
        vals[:,i] =X_choice.dot(beta_choice)[:,0]
        for j in range(N):
            variances[j,i]=(X_choice[j,:][np.newaxis,:]).dot(cov_choice).dot(X_choice[j,:][:,np.newaxis])
        #variances[:,i] =np.diagonal(X_choice.dot(cov_choice).dot(X_choice.T))
    return [vals,variances]
            

```

```python
def data_generate(X,n=1000):
    [N,p]=X.shape
    Xs =np.zeros((n,p),dtype=float)
    for i in range(p):
        Xs[:,i]=np.linspace(X[0,i],X[N-1,i],n);
    return Xs
    
            
            
```

```python
def split_dataset(data,train_nums):
    [N,p] =data.shape
    print('N ',N)
    index =np.array(range(N),dtype=int)
    choices =np.random.choice(a=index,size=train_nums,replace=False,p=None)
    data_train =data[choices,:].copy()
    selected =np.ones((N,),dtype=bool)
    selected[choices] =False
    data_test =data[index[selected],:].copy()
    return [data_train,data_test]
[data_train,data_test] =split_dataset(np.concatenate([aa_data,ao_data],axis=0),1000)
```

```python
def d(x,k,knots):
    a =np.power(x-knots[k],3)
    b =np.power(x-konts[-1],3)
    a =(int(a>=0))*a
    b =(int(b>=0))*b
    res =(a-b)/(knots[-1]-knots[k])
    return res
def N(x,k,knots):
    if(k==0):
        return x
    else:
        return d(x,k-1,knots)-d(x,knots.shape[0]-2,knots)
def generateH(freqs,knots_num=13):
    p =freqs.shape[0]
    pers=100*np.linspace(0,1,13)
    knots =np.percentile(freqs,pers)
    knots =knots.astype(np.int16)
    H =np.zeros((p,knots_num-1),dtype=float)
    for i in range(p):
        for k in  range(knots_num-1):
            H[i,k] =N(freqs[i],k,knots)
    return H

H =generateH(freqs=freqs,knots_num=13)
    
```

```python
class SmoothingSpline:
    def __init__(self,intercept=None,beta=None,N=None,Omega=None,S=None,knots=None,minval=None,maxval=None):
        
        self.intercept =intercept
        self.beta =beta
        self.N =N   #nature spline
        self.Omega =Omega #penality
        self.S =S   #smoother matrix
        self.knots =knots #knots
        self.minval =minval
        self.maxval =maxval
    
    def d(self,x,k,knots):
        eps =1e-72
        a =np.power(x-knots[k],3)
        b =np.power(x-knots[-1],3)
        a =(int(a>=0))*a
        b =(int(b>=0))*b
        
        return (a-b)/(eps+1.0*knots[-1]-knots[k])
    
    def nspline(self,x,k,knots):
        if(k==0):
            return 1
        elif(k==1):
            return x
        else:
            return self.d(x,k-2,knots) -self.d(x,knots.shape[0]-2,knots)
    
    def integral(self,j,k,knots,high,low):
        a =knots[j]+knots[k]
        b =knots[j]*knots[k]
        c =(knots[-1]-knots[j])*(knots[-1]-knots[k])
        high_val =(12*np.power(high,3)-18*a*np.square(high)+36*b*high)/c
        low_val  =(12*np.power(low,3)-18*a*np.square(low)+36*b*low)/c
        return high_val-low_val
    
    def integral_item(self,j,k,knots):
        if(j==0 or j==1 or k==0  or k==1):
            return 0
        else:
            low1 =max(knots[j],knots[k])
            high1 =knots[-1]
            
            low2 =knots[-2]
            high2 =knots[-1]
            
            A =self.integral(j-2,k-2,knots,high1,low1)
            B =self.integral(j-2,-2,knots,high2,low2)
            C =self.integral(k-2,-2,knots,high2,low2)
            D =self.integral(-2,-2,knots,high2,low2)
            
            return A-B-C+D
        
    def lbd_get(self,eigvals,df,tol=0.000001):
        N =eigvals.shape[0]
        #print('eigvals ',eigvals)
        #print('N = ',N)
        #print('eigvals>=0 ',np.count_nonzero(eigvals>=0))
        eigs =eigvals.copy()
        #fv =np.zeros((N,),dtype=float)
        #de =np.zeros((N,),dtype=float)
        lbd_last =12
        lbd_new =0  #lambda_new 选择不一样这里算出来的lambda不一样
        fv =df
        while(abs(fv)>tol or abs(lbd_new-lbd_last)>tol):
            #print('lbd_last ',lbd_last)
            #print('lbd_new ',lbd_new)
            lbd_last =lbd_new
            eps =1e-140
            fv =lbd_new*eigs
            fv =1.0/(1.0+fv+eps)
            de =np.square(fv.copy())
            de =-1.0*eigs*de
            de =np.sum(de)
            fv =np.sum(fv) -df
            #print('fv ',fv)
            lbd_new =lbd_new -(fv/(de+eps))
            if(lbd_new<0):
                lbd_new =np.random.rand()
            #print('update lbd_new ',lbd_new)
        #print('df= ',df)
        #print('df ',np.sum(1.0/(1+lbd_new*eigvals)))
        return lbd_new
    
    def fit(self,X,y,df=12):
        #m =X.shape[0]
        Xs =X.copy()
        ys =y.copy()
        minval =np.min(Xs)
        maxval =np.max(Xs)
        #Xs =(Xs-minval)/(maxval-minval)
        knots,mask =np.unique(Xs,return_index=True)
        #knots =knots[:-1]
        #mask  =mask[:-1]
        
        #knots =knots[0:20]
        #mask  =mask[0:20]
        Xs =knots
        ys =ys[mask]
        
        #if(Xs.shape[0]>49):
        #    index =np.zeros((Xs.shape[0],),dtype=bool)
        #    mm =int(Xs.shape[0]/25)
        #    #print('Xs.shape[0] ',Xs.shape[0])
        #    for i in range(Xs.shape[0]):
        #        if((i%mm)==0):
        #            index[i]=True
        #    #print('index ',index)
        #    Xs =Xs[index]
        #    ys =ys[index]
        #    knots =Xs
            
        
        #knots个数的增加会让矩阵变得奇异，小范围的还行。R语言中默认是
        #当n>49时，选择n^0.2个点作为knots，但是仅仅是出于计算量考虑,我这里数值不稳定
       # print('knots ',knots)
        
        m =knots.shape[0]
        #print('m ',m)
        
        N =np.zeros((m,m),dtype=float)
        Omega =np.zeros((m,m),dtype=float)
        
        for i in range(m):
            for j in range(m):
                N[i,j]=self.nspline(knots[i],j,knots)
                Omega[i,j]=self.integral_item(i,j,knots)
        
        temp0 =np.linalg.inv(N.T).dot(Omega).dot(np.linalg.inv(N))
        eig33 =np.linalg.eigvals(temp0)
        
        lbd =self.lbd_get(eig33,df)
        print('lambda ',lbd)
        #只用一种方式计算S，两个方法都用在一些数据上会出现不一致
        temp =np.linalg.inv(N.T.dot(N)+lbd*Omega).dot(N.T)
        S2 =np.linalg.inv(np.eye(N.shape[0],dtype=float)+lbd*temp0)
        beta =temp.dot(ys[:,np.newaxis])
        S =N.dot(temp)
        print('df1 ',np.sum(S.diagonal()))
        print('df2 ',np.sum(S2.diagonal()))
        #intercept =beta[0,0]
        #beta =beta[1:,0]
        
        #self.intercept =intercept
        self.minval =minval
        self.maxval =maxval
        self.beta =beta
        self.N =N
        self.Omega =Omega
        self.S =S
        self.knots=knots
    
    def predict(self,X):
        Xs =X.copy()
        #Xs =(Xs-self.minval)/(self.maxval-self.minval)
        m =Xs.shape[0]
        p =self.N.shape[0]
        N =np.zeros((m,p),dtype=float)
        for i in range(m):
            for j in range(p):
                N[i,j]=self.nspline(Xs[i],j,self.knots)
        
        y_predict =N.dot(self.beta)
        return y_predict[:,0]
    
        
        
    
    
        
    
        
        
```

```python
def generate_data(N=100):
    X =np.random.rand(N)
    noise =np.random.normal(loc=0.0,scale=1.0,size=100)
    y =np.zeros((N,),dtype=float)
    f =np.zeros((N,),dtype=float)
    for i in range(N):
        f[i] =(np.sin(12*(X[i]+0.2)))/(X[i]+0.2)
        y[i] =f[i]+noise[i]
    return [X,f,y]

```

```python
class Bsplines:
    def __init__(self,M=4,K=11,N=None,knots=None,points=None,basis=None):
        self.M=M
        self.K=K
        self.N=N
        self.knots=knots
        self.points=points
        self.basis=basis
    
    def B(self,i,m,x):
        eps=1e-5
        #print("i ",i," m ",m)
        if(m==1 and self.points[i]==self.points[i+1]):
            return 0.0
        elif(m==1 and self.points[i]<=x and x<self.points[i+1]):
            return 1.0
        elif(m==1):
            return 0.0
        else:
            #print('B(i,m-1,x) ',self.B(i,m-1,x))
            #print('B(i+1,m-1,x) ',self.B(i+1,m-1,x))
            #print('points[i+m-1] ',self.points[i+m-1],' points[i] ',self.points[i])
            #print('points[i+m] ',self.points[i+m],' poins[i+1] ',self.points[i+1])
            if(self.points[i+m-1]==self.points[i]):
                a=0.0
            else:
                a =self.B(i,m-1,x)*((x-self.points[i])/(self.points[i+m-1]-self.points[i]))
            if(self.points[i+m]==self.points[i+1]):
                b=0.0
            else:
                b =self.B(i+1,m-1,x)*((self.points[i+m]-x)/(self.points[i+m]-self.points[i+1]))
            #print('a ',a,' b ',b)
            #print('a+b ',a+b)
            return a+b
    
    def generate(self,M,knots,K,N):
        print(knots.shape)
        
        print(knots)
        points =np.zeros((K+2*M+1,),dtype=float)
        points[1:M+1]=knots[0]
        points[M+K+1:2*M+K+1]=knots[-1]
        points[M+1:M+K+1]=knots;
        
        self.M=M
        self.K=K
        self.N=N
        self.knots=knots
        self.points=points
        
        data=np.linspace(knots[0],knots[-1],N)
        
        basis =[]
        for m in range(1,M+1):
            X =np.zeros((N,K+2*M-m),dtype=float)
            for i in range(1,N+1):
                for k in range(1,K+2*M-m):
                    X[i-1,k-1]=self.B(k,m,data[i-1])
            basis.append(X)
        
       
        self.basis=basis
        return basis
        
```

```python
class wavelets:
    def __init__(self,V=None,J=9,N=512,basis=None,coefs=None,phi=None,psi=None,has=0):
        self.V=V
        self.J =J
        self.N =N
        self.data =data
        self.coefs=coefs
        self.phi=phi
        self.psi=psi
        self.has=0
        
    
    def haar(self,x):
        if(x>=0.0 and x<1.0):
            return 1.0
        else:
            return 0.0
    def db(self,x):
        if(self.has==0):
            wavelet=pywt.DiscreteContinuousWavelet('db8')
            [phi, psi, xx]=wavelet.wavefun()
            self.phi=phi
            self.psi=psi
            has=1
        if(x>=0 and x<15.0):
            return self.phi[int(self.phi.shape[0]*x/15.0)]
        else:
            return 0
        
    def father(self,x,fun):
        if(fun=='haar'):
            return self.haar(x)
        elif(fun=='db8'):
                return self.db(x)
    
    def mother(self,x,fun):
        return self.father(2*x,fun)-self.father(2*x -1.0,fun)
    
    def scale(self,j,k,fun,x):
        if(j==0):
            return self.father(x,fun)
        else:
            return np.power(2,j/2)*self.father(np.power(2,j)*x-k,fun)
    def detail(self,j,k,fun,x):
        if(j==0):
            return self.mother(x,fun)
        else:
            return np.power(2,j/2)*self.mother(np.power(2,j)*x-k,fun)
        
    def basis(self,V,J,y,fun):
        ys =y.copy();
        N =(1<<J)
        self.J=J
        self.N=N
        self.V=V
        data =np.zeros((N,N),dtype=float)
        
        X =np.array(range(0,N),dtype=float)/N
        
        index =0
        for j in range(J):
            for k in range((1<<j)):
                for i in range(N):                
                    if(index<(1<<V)):
                        data[i,index]=self.scale(j=V,k=k,fun=fun,x=X[i])
                    else:
                        data[i,index]=self.detail(j=j,k=k,fun=fun,x=X[i])
                index =index+1
        
        self.data=data
        
        coefs =data.T.dot(y[:,np.newaxis])
        self.coefs=coefs
        
        """
        v=coefs[0]
        arr =coefs[(1<<(J-1)):-1]
        mid =np.median(arr)
        arr =arr -mid
        std =(np.sqrt(N)*np.median(arr))/0.6745
        lmd =std*np.sqrt(2*np.log(N)/N)
        mask =(coefs>=lmd)
        coefs[mask]=coefs[mask]-lmd
        coefs[~mask]=0
        coefs[0]=v
        """
        return [data,coefs]
    
        


                
                
    
        
        
```
